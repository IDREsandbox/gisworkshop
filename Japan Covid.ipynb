```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go

import ipywidgets as widgets
from ipywidgets import interact, interact_manual
import json
from pandas import json_normalize  
import requests
```

```python

import requests
import pandas as pd
r = requests.get('https://data.covid19japan.com/summary/latest.json')
x = r.json()
date = []
prefecture = []
dailyConfirmedCount = []
dailyDeceasedCount = []
for i in x['prefectures']:
    for j in range(len(i['dailyConfirmedCount'])):
        date.append(pd.to_datetime(i['dailyConfirmedStartDate']) + pd.DateOffset(j))
        prefecture.append(i['name'])
        dailyConfirmedCount.append(i['dailyConfirmedCount'][j])
        dailyDeceasedCount.append(i['dailyDeceasedCount'][j])
d = {'date':date,'prefecture':prefecture,'dailyConfirmedCount':dailyConfirmedCount,'dailyDeceasedCount':dailyDeceasedCount}
df = pd.DataFrame(data=d)
df
```

```python
fig = px.line(df,x='date',y='dailyConfirmedCount',facet_col='prefecture',facet_col_wrap=4,height=4000)
fig.update_yaxes(
    matches=None,
    showticklabels=False,
    showgrid=False,
    title_text='y'
) 

fig.update_layout(
    paper_bgcolor="LightSteelBlue",
)

fig.show()
```
