```python
date = []
confirmed_cases = []
place = []
counter = 1
for j in range(0,len(jh)):
    thisrow = jh.iloc[j]
#     print(thisrow[100])
    if pd.isnull(thisrow['state']):
        thisplace= str(thisrow['country'])
    else:
        thisplace = str(thisrow['state']) + ', ' + str(thisrow['country'])
#     print(place)
    for i in range(4,len(jh.columns)):
        date.append(jh.columns[i])
        confirmed_cases.append(thisrow[i])
        place.append(thisplace)
```

```python
fig = go.Figure(data=[go.Table(
    cells=dict(values=[date,place,confirmed_cases]))
])
# fig.show()

fig = px.line(x=date,
             y=confirmed_cases,
             color=place)
fig.show()
```

```python
data = {'date': date,
       'confirmed_cases': confirmed_cases,
       'place': place}
df = pd.DataFrame(data=data)
```

```python
px.line(df,x='date',y='confirmed_cases',color='place')
# px.line(df,x='date',y='confirmed_cases',color='place',facet_col='date',facet_col_wrap=4)
```

```python
#group by country, find the largest value epr country, sort it by most confirmed cases
top_countries = df.groupby('place').apply(lambda x: x.nlargest(1, 'confirmed_cases')).reset_index(drop=True).sort_values(by=["confirmed_cases"],ascending=False)

howmany = 20
top_countries = top_countries.head(howmany)
top_countries
top_countries_list = top_countries.place

top_countries_list = list(top_countries_list)
```

```python
# date not sorted correctly. let's try again
from datetime import datetime
df["date"] = pd.to_datetime(df["date"], format="%m/%d/%y")\
               .dt.strftime('%Y-%m')
```

```python
# Create a data frame for just the top countries
top_countries_df=df.query("place == @top_countries_list").sort_values(by=["date"])
top_countries_df

```

```python
fig=px.line(top_countries_df,
        x='date',
        y='confirmed_cases',
        facet_col='place',
        facet_col_wrap=5,
        height=1000)

# individual scale for each y axis
fig.update_yaxes(matches=None,
                 showticklabels=False,
                 showgrid=False,
                 title_text='') 
fig.update_xaxes( title_text='') 
# title: show only the country name
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]+'...'))

fig.update_layout(title_text="Covid-19 Confirmed Cases by Country")
fig.show()
```
